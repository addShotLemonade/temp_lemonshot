```python
import pandas as pd
import unicodedata

# CSV 로드
celeb_df = pd.read_csv('celeb_data.csv')
celeb_df = celeb_df.rename(columns={"Celebrity Name": "celeb_name"})

# 정규화 함수
def normalize_korean(text):
    return unicodedata.normalize("NFC", text)

# 정규화 적용
celeb_df['celeb_name'] = celeb_df['celeb_name'].apply(normalize_korean)

# 여성 / 남성 분리
female_df = celeb_df.iloc[:86]
male_df = celeb_df.iloc[86:]

# TypeScript 코드 생성
def to_ts_array(name: str, df: pd.DataFrame, gender: str) -> str:
    entries = [
        f"  {{ name: '{row}', image: '/images/{gender}/{row}.png' }}"
        for row in df['celeb_name']
    ]
    return f"const {name}: Candidate[] = [\n" + ",\n".join(entries) + "\n];"

female_ts = to_ts_array("femaleCandidates", female_df, "female")
male_ts = to_ts_array("maleCandidates", male_df, "male")

# 저장
with open("candidates.ts", "w", encoding="utf-8") as f:
    f.write(male_ts + "\n\n" + female_ts)

print("✅ candidates.ts 파일 생성 완료!")
```
